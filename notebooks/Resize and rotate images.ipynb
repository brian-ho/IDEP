```python
import os
import pandas as pd
import scipy.misc
import PIL
from PIL import Image

common_dir = './20171007'

# Get images from folder
images = [os.path.join(common_dir, f) for f in os.listdir(common_dir)
         if os.path.isfile(os.path.join(common_dir, f)) and not f.startswith('.')]

print len(images)
```

```python
# For each image path, rotate and resize
for path in images:

    img = Image.open(path)
    img = img.resize((256,256), PIL.Image.ANTIALIAS)
    img = img.rotate(-90)
    
    # Save to folders based on label
    name = path[-12:]
    img.save('./label_2/'+name)
```
