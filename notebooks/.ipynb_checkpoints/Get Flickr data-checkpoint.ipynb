```python
def get_photo_data(img_id):
    start = time.time()
    try:
        payload = {
            'photo_id' : img_id,
            'method' : 'flickr.photos.getInfo',
            'api_key' : FLICKR_API_KEY,
        }

        r = requests.get('https://api.flickr.com/services/rest/', params = payload)
        
    except:
        return
    
    data = yahoo.data(ElementTree.fromstring(r.content))
    soup = BeautifulSoup(data['rsp']['photo']['description'], "lxml")
    
    for b in soup.find_all('b'):
        kl.ix[str(img_id), b.text] = b.next_sibling.strip()
        
    kl.ix[str(img_id),'Check'] = 1;
    
    end = time.time()
    delay = 1.0 - (end-start)
                   
    if delay > 0:
        time.sleep(delay)
```
