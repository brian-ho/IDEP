```python
def detect_landmarks(path):
    """Detects landmarks in the file."""
    client = vision.ImageAnnotatorClient()

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image = types.Image(content=content)

    response = client.landmark_detection(image=image)
    landmarks = response.landmark_annotations
    return landmarks
    print('Landmarks:')

    for landmark in landmarks:
        print(landmark.description)
        for location in landmark.locations:
            lat_lng = location.lat_lng
            print('Latitude'.format(lat_lng.latitude))
            print('Longitude'.format(lat_lng.longitude))
```

```python
def detect_properties(path):
    color_ = {}
#     print path
    
    """Detects image properties in the file."""
    client = vision.ImageAnnotatorClient()

    with io.open(path, 'rb') as image_file:
        content = image_file.read()

    image = types.Image(content=content)

    response = client.image_properties(image=image)
    props = response.image_properties_annotation
#     print('Properties:')
#     print props
    for color in props.dominant_colors.colors:
        color_['r']=color.color.red
        color_['g']=color.color.green
        color_['b']=color.color.blue
        color_['frac']=color.pixel_fraction
        color_['score']=color.score
            
    return color_
#         print('fraction: {}'.format(color.pixel_fraction))
#         print('\tr: {}'.format(color.color.red))
#         print('\tg: {}'.format(color.color.green))
#         print('\tb: {}'.format(color.color.blue))
#         print('\ta: {}'.format(color.score))
```

```python
colors = []
for i,image_ in enumerate(images):
    if i % 100 == 0:
        print "%i of %i" % (i, len(images))
        
    color=detect_properties('/Volumes/bho_250GB/IDEP/For Web Small/%s' % image_)
    colors.append(color)
```

```python
pickle.dump( colors, open( "./GoogleCloudVisionColors.p", "wb" ) )
```

```python
image_labels = []

for i, image_path in enumerate(images):
    if i % 100 == 0:
        print "%i of %i" % (i, len(images))
        
    # Instantiates a client
    client = vision.ImageAnnotatorClient()

    # The name of the image file to annotate
    file_name = os.path.join('/Volumes/bho_250GB/IDEP/For Web Small/%s' % image_path)

    # Loads the image into memory
    with io.open(file_name, 'rb') as image_file:
        content = image_file.read()

    image = types.Image(content=content)

    # Performs label detection on the image file
    response = client.label_detection(image=image)
    labels = response.label_annotations
    
    img_labels = {'image': image_path}
#     print('%s labels:' % image_path)
    for label in labels:
#         print(label.description, label.score)
        img_labels[label.description] = (label.score, label.mid)
    
    image_labels.append(img_labels)
    time.sleep(0.1)
```
