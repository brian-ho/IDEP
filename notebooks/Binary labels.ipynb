```python
import pandas as pd
import numpy as np
import shutil
import requests
import urllib
from PIL import Image
from IPython.display import Image as Image2
from StringIO import StringIO
import os
```

```python
df = pd.read_csv('../data/kl_geocode_label_vector2.csv')

ids = df.id.values
print ids
df.head()
```

```python
def make_split(i):
    if i < 0.7:
        return 0
    elif i < 0.9:
        return 1
    else:
        return 2
```

```python
for element in ['node','district','edge','path','landmark']:
    df['%s_msk' % element] = np.random.rand(len(df))
    df['%s_msk' % element] =  df['%s_msk' % element].apply(lambda x: make_split(x))

df.head()
```

```python
joined = []
lookup = {}

for path, _, f in os.walk('../labels/all'):
    joined = f

for image_name in joined:
    lookup[int(image_name[:10])] = image_name
```

```python
for i, image in df.iterrows():
    print '%i of %i' % (i, len(joined))
    src = '../labels/all/%s' % (lookup[image.id])
    
    for element in ['path']:# ['node','district','edge','path','landmark']:
        if image['%s_msk' % element] == 0:
            dst = '../labels/%s/train/%i/' % (element, image[element])
        elif image['%s_msk' % element] == 1:
            dst = '../labels/%s/test/%i/' % (element, image[element])
        else:
            dst = '../labels/%s/holdout/%i/' % (element, image[element])
#         print src, dst
        shutil.copy(src, dst)
```
