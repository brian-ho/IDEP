```python
import pandas as pd
import requests
import time
```

```python
data = pd.read_csv('data/kl_full.csv')
```

```python
data['query_result'] = 0

for i in range(5):
    data['format%i' % i] = ''
    data['lat%i' % i] = ''
    data['lng%i' % i] = ''
    
data.head()
```

```python
def geocode(row, index):
    
    print "Checking image %i" % (row.id),
    
    params = {
        'address' : row['query'],
    }

    r = requests.get('http://maps.googleapis.com/maps/api/geocode/json?', params)
    
    if r.json()['status'] == 'ZERO_RESULTS':
        print '... 0'
        return
        
    else:
        data.loc[index, 'query_result'] = 1
        
        results = r.json()['results']
        print '... %i' % (len(results))
        
        for i, result in enumerate(results):
            data.loc[index, 'format%i' % i] = result['formatted_address']
            data.loc[index, 'lat%i' % i] = result['geometry']['location']['lat']
            data.loc[index, 'lng%i' % i] = result['geometry']['location']['lng']
            
            if i == 4:
                return
        
        return
```

```python
for index, row in data.iterrows():
    geocode(row, index)
    time.sleep(0.02)
```
